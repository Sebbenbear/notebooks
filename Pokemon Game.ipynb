```python
import pandas as pd
import seaborn as sns
import numpy as np
import requests
import json
```

```python
r = requests.get('http://pokeapi.co/api/v2/pokemon/1/')
response_json = r.json()

print(json.dumps(response_json, indent=4, sort_keys=False))
```

```python
pokemon_name = response_json['forms'][0]['name'].title()

print("You got {}!".format(pokemon_name))
```

```python
question = "Which pokemon do you choose? (1) Bulbasaur, (2) Charmander, (3) Squirtle}"

answer = int(input())

pokemon_number = 0

# Go back and choose a name for charmander:

if answer == 1:
    pokemon_number = 1
elif answer == 2:
    pokemon_number = 4
elif answer == 3:
    pokemon_number = 7
else:
    answer = False
    
if answer:
    response_json = requests.get('http://pokeapi.co/api/v2/pokemon/{}/'.format(pokemon_number)).json()
    pokemon_name = response_json['forms'][0]['name'].title()
    print("You got {}!".format(pokemon_name))
    
```

```python
pokemon_move = response_json['moves'][0]['move']['name']
print("{} knows {}!".format(pokemon_name, pokemon_move))
```

```python
pokemon_moves = response_json['moves']

squirtle_moves = []
for i in range(len(pokemon_moves)):
    squirtle_moves.append(pokemon_moves[i]['move']['name'])
    
print(squirtle_moves)
    
#print(json.dumps(pokemon_moves, indent=4))
```

```python
stats_json = response_json['stats']
#pokemon_name = response_json['forms'][0]['name'].title()
print(response_json['name'].title())

base_stats = {}
for stat in stats_json:
    stat_name = stat['stat']['name']
    base_stat = stat['base_stat']
    print("{0:15}  {1}".format(stat_name, base_stat))
    base_stats[stat_name] = base_stat
    
print(base_stats)

#print(json.dumps(stats_json, indent=4))
```
